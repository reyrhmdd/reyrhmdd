MEMBUAT VISUALISASI GARIS KEMISKINAN DI PROVINSI LAMPUNG

```python
import pandas as pd
import matplotlib.pyplot as plt

# Memuat dataset
data = pd.read_csv('/content/Garis Kemiskinan Kapita Perbulan (Rp_Kap_Bulan), 2023-2024.csv')

# Membersihkan dan menyiapkan data
data_bersih = data.rename(columns={'Wilayah': 'Wilayah', 'Unnamed: 1': '2023', 'Unnamed: 2': '2024'})
data_bersih = data_bersih.drop([0, 1])  # Menghapus baris metadata
data_bersih['2023'] = pd.to_numeric(data_bersih['2023'], errors='coerce')
data_bersih['2024'] = pd.to_numeric(data_bersih['2024'], errors='coerce')
data_bersih = data_bersih.dropna(subset=['Wilayah'])

# Boxplot untuk tahun 2023 dan 2024
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
data_bersih[['2023', '2024']].boxplot()
plt.title('Boxplot Garis Kemiskinan (2023 vs 2024)')
plt.ylabel('Garis Kemiskinan (Rp/Kapita/Bulan)')

# Barplot perbandingan garis kemiskinan tahun 2023 dan 2024
plt.subplot(1, 3, 2)
data_bersih.plot(x='Wilayah', y=['2023', '2024'], kind='bar', ax=plt.gca(), color=['skyblue', 'lightgreen'])
plt.title('Barplot Garis Kemiskinan (2023 vs 2024)')
plt.ylabel('Garis Kemiskinan (Rp/Kapita/Bulan)')
plt.xticks(rotation=90)

# Scatter plot perbandingan garis kemiskinan tahun 2023 dan 2024
plt.subplot(1, 3, 3)
plt.scatter(data_bersih['2023'], data_bersih['2024'], color='purple')
plt.title('Scatter Plot Garis Kemiskinan (2023 vs 2024)')
plt.xlabel('Garis Kemiskinan 2023 (Rp)')
plt.ylabel('Garis Kemiskinan 2024 (Rp)')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membuat heatmap
plt.figure(figsize=(10, 8))

# Membuat data untuk heatmap
heatmap_data = data_bersih.set_index('Wilayah')[['2023', '2024']]

# Menghasilkan heatmap
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', linewidths=0.5, fmt='.0f')

plt.title('Heatmap Garis Kemiskinan (2023 vs 2024)')
plt.xlabel('Tahun')
plt.ylabel('Wilayah')

plt.tight_layout()
plt.show()
```
